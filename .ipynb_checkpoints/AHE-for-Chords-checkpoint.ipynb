```python
import os
import jams
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import jams.display
from tqdm import tqdm_notebook as tqdm
import pandas as pd
from joblib import Parallel, delayed
import mir_eval
import pumpp
from copy import deepcopy
```

```python
def getid(fname):
    return os.path.splitext(os.path.basename(fname))[0]
```

```python
def issame(labs1, labs2):
    
    # Hash the strings
    h1 = [hash(_) for _ in labs1]
    h2 = [hash(_) for _ in labs2]
    
    # Build the segment label agreement matrix
    a1 = np.equal.outer(h1, h1)
    a2 = np.equal.outer(h2, h2)
    
    # Labelings are the same if the segment label agreements are identical
    return np.all(a1 == a2)
```

### Functions to create the hierarchies

In this section, we define two functions `chord_to_hier` and `hier_pruned` that create hierarchical representations for chord annotations. We then define a third function `mplot` that creates visual representations for these hierarchical representations. 

The first function creates a hierarchical representation for a given sequence of chord annotations. The resulting hierarchy has six levels: 
1. Just the simplified root
2. The chord annotation with the simplified root specified to its maj/min designation
3. The chord annotation with the root simplified specified to its aug/dim designation
4. The chord annotation with the root simplified and the inversions dropped. This level is where designations with maj6/min6 and 7/min7/maj7/dim7/hdim7/minmaj7 are preserved. 
5. The chord annotation with only the root simplified (meaning that the inversions are preserved)
6. The original chord annotation

Noting that not all annotations make use of the full chord vocabulary, the second function `hier_pruned` simplifies the hierarchy resulting from `chord_to_hier` such that the $n^{th}$ level is only included if it contains additional information from the $(n-1)$ level. For example, level 3 is only included if there are any chords that carry the aug/dim designations; otherwise levels 2 and 3 contain the same information. 

```python
def rootX(label, root):
    
    if label == 'X':
        return f'{root}:{label}'
        
    return label

def chord_to_hier(ann):
    """Creates a hierarchy from chord information"""
    
    # Do the isolations for each observed chord
    C3 = pumpp.task.ChordTagTransformer(vocab='3')
    C35 = pumpp.task.ChordTagTransformer(vocab='35')
    C3567s = pumpp.task.ChordTagTransformer(vocab='3567s')
    
    # Create object to place the annotations into
    ann_h = jams.Annotation('multi_segment',
                            time=ann.time,
                            duration=ann.duration,
                            annotation_metadata=ann.annotation_metadata)
    
    for obs in ann:
        og_chord = obs.value
        if ':' in og_chord:
            root, dressing = og_chord.split(':', 2)
            # Isolate the enharmonically unique root
            root = C3567s.simplify(root)
            root = root[:root.index(':')] 
            root_change = root + ':' + dressing
        else:
            # Isolate the enharmonically unique root
            root = C3567s.simplify(og_chord)
            if ':' in root:
                root = root[:root.index(':')] 
            root_change = root

        # Make first level with the root only 
        ann_h.append(time=obs.time, duration=obs.duration,
                    value=dict(level=1, label=root),
                    confidence=obs.confidence)
        
        # Add the maj/min designation to the root in level 2
        ann_h.append(time=obs.time, duration=obs.duration,
                    value=dict(level=2, label=rootX(C3.simplify(obs.value), root)),
                    confidence=obs.confidence)
        
        # Add aug/dim to the above
        ann_h.append(time=obs.time, duration=obs.duration,
                    value=dict(level=3, label=rootX(C35.simplify(obs.value), root)),
                    confidence=obs.confidence)
        
        # Above plus maj6/min6 & 7/min7/maj7/dim7/hdim7/minmaj7
        # Only inversions dropped here
        ann_h.append(time=obs.time, duration=obs.duration,
                    value=dict(level=4, label=rootX(C3567s.simplify(obs.value), root)),
                    confidence=obs.confidence)
        
        # Just the root changed (but inversions kept)
        ann_h.append(time=obs.time, duration=obs.duration,
                    value=dict(level=5, label=root_change),
                    confidence=obs.confidence)
        
        # Original string 
        ann_h.append(time=obs.time, duration=obs.duration,
                    value=dict(level=6, label=obs.value),
                    confidence=obs.confidence)
    return ann_h
```

```python
def hier_pruned(_ann_h):
    """Prune the levels in the hierarchy that are the same as the previous one. """
    
    # Make a copy of the hierarchy
    ann_ph = deepcopy(_ann_h)
    
    # Flatten the annotations and zip the pairs of information together
    h_intervals, h_labels = jams.eval.hierarchy_flatten(ann_ph)
    ann_z = zip(h_intervals, h_labels)
    ann_z = list(ann_z)
    
    # Remove the old segment data
    ann_ph.pop_data()
    
    # Determine the height of the input hierarchy:
    max_level = len(ann_z)    
    
    
    # Initialize the pruned heirarchy
    
    # Start with the lowest level
    level = 0
    # Create first part of the hierarchy
    ints, labs = ann_z[0]
    for ival, label in zip(ints, labs):
        ann_ph.append(time=ival[0], duration=ival[1]-ival[0], value=dict(label=label, level=level))
        
    # Advanced the level marker
    level += 1
    # Set previous ints and labs as new variables
    prev_ints = ints
    prev_labs = labs
    
    
    # For each remaining level in the hierarchy, check that the level is adding something new 
    #     that is not contained in the previous level
    for i in range(1, max_level):
        
        # Extract the new level
        ints, labs = ann_z[i]
        
        # Only add if this level is different than the previous one: 
        if not issame(prev_labs, labs):
            for ival, label in zip(ints, labs):
                ann_ph.append(time=ival[0], duration=ival[1]-ival[0], value=dict(label=label, level=level))
            
            # If a level was added, then update comparison point and update the level marker
            # Advanced the level marker
            level += 1
            # Set previous ints and labs as new variables
            prev_ints = ints
            prev_labs = labs
    
    #print(level)
    return ann_ph
```

```python
def mplot(test_ann_h, **kwargs):
    """Visualize a resulting hierarchy by adding leading spaces to the labels on each level. """
    hier_ints, hier_labs = jams.eval.hierarchy_flatten(test_ann_h)
    
    for n in range(len(hier_labs)):
        for m in range(len(hier_labs[n])):
            hier_labs[n][m] = ' ' * n + hier_labs[n][m]
            
    return mir_eval.display.hierarchy(hier_ints, hier_labs, **kwargs)
```

```python
from tqdm.notebook import tqdm
import itertools
from pathlib import Path
```

```python
chord_ref_loc = '/home/bmcfee/git/ismir2017_chords/data/references_v2/'
chord_est_loc = '/home/bmcfee/git/ismir2017_chords/data/model_deep_struct_aug/'

jamses_ref = jams.util.find_with_extension(chord_ref_loc, 'jams')
jamses_est = jams.util.find_with_extension(chord_est_loc, 'jams')
```

```python
def fix_annotation(ann):
    # Fix chord annotations to avoid interval overlaps
    # This used to work fine in mir_eval, but the introduction of segmentation metrics in 0.6 broke things

    intervals, values = ann.to_interval_values()
    ann.pop_data()
    last = 0
    
    for ival, val in zip(intervals, values):
        t = max(last, ival[0])
        if ival[1] > t:
            ann.append(time=t, duration=ival[1] - t, value=val, confidence=None)
        last = ival[1]
    

def jamses_merged(jref, jest, outputdir='/home/bmcfee/git/tismir2020-hierarchy/data/merged'):
    
    for j1, j2 in tqdm(zip(jref, jest)):
        if Path(j1).name != Path(j2).name:
            raise Exception('Naming mismatch: {}, {}'.format(Path(j1).name, Path(j2).name))
            
        J = jams.load(j1, validate=False)
        J2 = jams.load(j2, validate=False)
        
        # Smash the two annotations together into one jams file
        J.annotations.extend(J2.annotations)
        
        # Make sure we only have one annotation from each side
        if len(J.annotations) != 2:
            continue
            
        # Update their metadata so we can bookkeep later
        J.annotations[0].annotation_metadata.annotator.name = 'Reference'
        J.annotations[1].annotation_metadata.annotator.name = 'Estimate'
        
        # Correct the annotations so that there's no oevrlapping intervals
        for ann in J.annotations['chord']:
            fix_annotation(ann)
            
        J.save(str(Path(outputdir) / Path(j1).name), strict=False)
```

### Defining comparison functions

We define functions for comparing the automatic hierarchy expansions in a few ways. 

1. The function `compare_hierarchies` compares the L-measure for the automatic hierarchy expansions of chords to the typical chord metrics (such as tetrads, root, and mirex)
2. The function `compare_flat` compares the L-measure for the flat chord annotations to the automatic hierarchy expansions of the same annotations. 
   - The helper function `flat_to_hierarchy` converts the flat annotation to one with the `multi-segment` designation. This function does not add any levels to the flat annotation. This change is needed for comparing the flat annotations to the hierarchical versions using the `jams` package. 

```python
FRAME_SIZE = 0.1

def compare_hierarchies(jamfile, pruned=True):
    
    J = jams.load(jamfile, validate=False)
    
    if len(J.annotations['chord']) != 2:
        return
    
    data = []
    index = []
    
    # iterate of pairs of chord annotations
    for a1, a2 in itertools.combinations(J.annotations['chord'], 2):
    
        # Compute standard chord metrics on the flat annotations
        losses = jams.eval.chord(a1, a2, frame_size=FRAME_SIZE)
        
        # Creating the hierachies from the chord annotations
        _e1 = chord_to_hier(a1)
        if pruned:
            e1 = hier_pruned(_e1)
        else:
            e1 = _e1
        
        _e2 = chord_to_hier(a2)
        if pruned:
            e2 = hier_pruned(_e2)
        else:
            e2 = _e2
    
        # Comparing hierarchical annotations to hierarchical annotations 
        losses_exp = jams.eval.hierarchy(e1, e2, frame_size=FRAME_SIZE)
        
        # Save the results of the standard chord metrics as well as the hierarchical comparisons
        losses.update({'Expanded {}'.format(_): losses_exp[_] for _ in losses_exp})
    
        losses = {_: losses[_] for _ in losses if 'T-' not in _}
        losses['ref_name'] = a1.annotation_metadata.annotator.name
        losses['est_name'] = a2.annotation_metadata.annotator.name
        
        data.append(losses)
        index.append(jams.util.filebase(jamfile))
    
    return pd.DataFrame(data, index=index)
```

```python
FRAME_SIZE = 0.1

def flat_to_hierarchy(ann, full=False):
    """Convert the flat annotation to a 'multi_segment' one. This function
    does not add any levels to the flat annotation."""
    
    ann_h = deepcopy(ann)
    ann_h.namespace = 'multi_segment'
    
    for obs in ann_h.pop_data():
        ann_h.append(time=obs.time, duration=obs.duration, confidence=obs.confidence,
                     value=dict(level=0, label=obs.value))
    return ann_h

def compare_flat(jamfile, key, pruned=True):
    
    J = jams.load(jamfile, validate=False)
    
    if len(J.annotations[key]) != 2:
        return
    
    data = []
    index = []
    
    for a1, a2 in itertools.combinations(J.annotations[key], 2):
    
        # Comparing flat annotations to flat annotations 
        f1 = flat_to_hierarchy(a1)
        f2 = flat_to_hierarchy(a2)
        losses = jams.eval.hierarchy(f1, f2, frame_size=FRAME_SIZE)
    
        # Creating the hierachies from the chord annotations
        _e1 = chord_to_hier(a1)
        if pruned:
            e1 = hier_pruned(_e1)
        else:
            e1 = _e1
        
        _e2 = chord_to_hier(a2)
        if pruned:
            e2 = hier_pruned(_e2)
        else:
            e2 = _e2
    
        # Comparing hierarchical annotations to hierarchical annotations 
        losses_exp = jams.eval.hierarchy(e1, e2, frame_size=FRAME_SIZE)
    
        # Save the results of both the flat and hierarchical comparisons
        losses.update({'Expanded {}'.format(_): losses_exp[_] for _ in losses_exp})
    
        losses = {_: losses[_] for _ in losses if 'T-' not in _}
        
        losses['ref_name'] = a1.annotation_metadata.annotator.name
        losses['est_name'] = a2.annotation_metadata.annotator.name
        
        data.append(losses)
        index.append(jams.util.filebase(jamfile))
    
    return pd.DataFrame(data, index=index)
```

## Experiments on Chord Annotations

In this section, we begin by applying the two above comparison functions to the chord annotations. First we compare the L-measures of the automatic hierarchy expansions for the chord annotations against 15 standard chord evaluation metrics. Then we compare the flat annotations against the expanded hierarchical annotations. In the remainder of the notebook, we create visual and numerical summaries to investigate the results of these comparison functions.
```python
def display_annotation(idx, ax=None):
    J = jams.load(str(Path('/home/bmcfee/git/tismir2020-hierarchy/data/merged/') / (str(idx) + '.jams')))
    
    if ax is None:
        ax = plt.gca()
        
    jams.display.display(J.annotations['chord', 0], meta=False, label='Reference', alpha=0.5, ax=ax)
    jams.display.display(J.annotations['chord', 1], meta=False, label='Estimate', alpha=0.5, ax=ax)
    
    ax.legend(loc='upper left')
```
